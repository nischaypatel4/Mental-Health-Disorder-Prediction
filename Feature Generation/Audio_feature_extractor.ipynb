Successfull eGeMAPS extraction

```python
# from ".mp4" to ".wav"

!ffmpeg -i "/content/sample.mp4" -vn "/content/sample_AUDIO.wav"
audio_path = '/content/sample_AUDIO.wav'

smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors
)

df_egemaps = smile.process_file(audio_path)

df_egemaps.reset_index(drop=False, inplace=True)
df_egemaps.to_csv('egemaps_from_python_opensmile.csv', index=False)
print("---------------eGeMAPS successfully extracted--------------")
df_egemaps.head()
```

MFCC Extraction Code

```python
smile_ll = opensmile.Smile(
    feature_set=opensmile.FeatureSet.ComParE_2016,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors
)
smile_deltas = opensmile.Smile(
    feature_set=opensmile.FeatureSet.ComParE_2016,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors_Deltas
)

df_ll = smile_ll.process_file(audio_path)
df_delta = smile_deltas.process_file(audio_path)

mfcc = df_ll.filter(like='mfcc_sma')
mfcc_delta = df_delta.filter(like='mfcc_sma_de')
mfcc_delta_delta = pd.DataFrame(np.diff(mfcc_delta, axis=0, prepend=np.nan), columns=[col.replace('_de', '_de_de') for col in mfcc_delta.columns])
```

### Feature Extraction

Emotion recognition from speech requires the extraction of relevant features from the audio signal. In this work, we utilized the **eGeMAPSv02** feature set, a widely used standard in paralinguistic analysis, extracted using the `opensmile` Python library.

The audio data, initially in MP4 format, was first converted to WAV format using the `ffmpeg` tool to be compatible with `opensmile`. The feature extraction process was then performed on the resulting WAV file (`/content/sample_AUDIO.wav`) using the `opensmile.Smile` class, specifying the `eGeMAPSv02` feature set and `LowLevelDescriptors` feature level.

The extracted features, consisting of a set of 23 acoustic parameters, were stored in a pandas DataFrame. This DataFrame includes information about the file path, the start and end time of each frame, and the extracted feature values for that frame. For further processing, the hierarchical index of the DataFrame was reset, and the resulting DataFrame was saved to a CSV file named `egemaps_from_python_opensmile.csv`.
